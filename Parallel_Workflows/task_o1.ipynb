```python
from langgraph.graph import StateGraph, START,END
from langchain_huggingface import HuggingFacePipeline,ChatHuggingFace
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from typing import TypedDict
```

```python
class StateGraph(TypedDict):
    pass
```

```python
class TaskO1(StateGraph):
    def __init__(self):
        super().__init__()
        self.add_node(START, HuggingFacePipeline.from_model_id(model_id="google/flan-t5-large", task="text2text-generation"), name="TaskO1_Start")
        self.add_node(END, ChatHuggingFace.from_model_id(model_id="gpt2", task="text-generation"), name="TaskO1_End")
        self.add_edge(START, END)
```

```python
graph=StateGraph()
```
